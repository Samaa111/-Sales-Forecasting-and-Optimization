The promotion period in late 2024 resulted in an extreme, short-term spike, generating daily sales of over $\$175,000$far exceeding any regular sales day. **Line Graph: Average Order Value by Day**

The overall trend shows that the **Average Order Value (AOV)** has stabilized over time, fluctuating around the $\$150$ mark for most of 2024 and 2025.

**Line Graph: Repeated Customers Over Time**

This chart shows the number of repeat customers from October 2023 to October 2025. The number of repeat customers showed steady growth throughout 2024, accelerating in the second half of the year to a peak of nearly 1,200 customers around November 2024. After a volatile but strong period in the first half of 2025, the count began to decline. The most critical observation is the sudden, sharp drop to zero in the final months of the timeline (late 2025), which likely indicates an abrupt end to the data collection or a significant business event.

```python
repeat_over_time = (
    df.loc[:, ['customer_unique_id', 'order_date']]
      .sort_values(['customer_unique_id', 'order_date'])
      .assign(prior_orders=lambda x: x.groupby('customer_unique_id').cumcount())  # repeats logic
      .assign(is_repeat=lambda x: x['prior_orders'] > 0)
      .assign(YearMonth=lambda x: x['order_date'].dt.to_period('M').dt.to_timestamp())
      .query('is_repeat')
      .groupby('YearMonth', as_index=False)['customer_unique_id'].nunique()
      .rename(columns={'customer_unique_id': 'repeat_customers'})
)


fig = px.line(
    repeat_over_time,
    x='YearMonth',
    y='repeat_customers',
    markers=True,
    title='Repeated Customers Over Time'
)
fig.show()
```
